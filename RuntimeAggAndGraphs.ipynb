```python
max_l = {}
max_r = {}

with open('src/result.txt', 'r') as file:
    lines = file.readlines()

for line in lines:
    parts = line.split()
    if parts[0] == 't':
        total = parts[1]
    if parts[0] == 'f':
        first = parts[1]
    if len(parts) >= 3:
        bin_count = int(parts[0])
        half = parts[1]
        value = float(parts[2])
        if half == 'l':
            if bin_count not in max_l or value > max_l[bin_count]:
                max_l[bin_count] = value
        elif half == 'r':
            if bin_count not in max_r or value > max_r[bin_count]:
                max_r[bin_count] = value

# Sum the largest values for 'l' and 'r'
sum_l = sum(max_l.values())
sum_r = sum(max_r.values())
other = float(total)-float(first)-max([sum_l,sum_r])
print(f"T {total}")  
print(f"F {first}")
print(f"L {sum_l}")
print(f"R {sum_r}")
print(f"O {other}")
```
